```python
#--- Librerias
import os
import cv2
from matplotlib import pyplot as plt
from distutils.dir_util import copy_tree
```

```python
#--- COPIANDO DIRECTORIOS Y FICHEROS.

DEFAULT_PATH = './SFEW' #-Airan Main Path
# (71, 90), (35, 45), (17, 22)
PATH_71x90 = './SFEW_71x90'
PATH_35x45 = './SFEW_35x45'
PATH_17x22 = './SFEW_17x22'

print("--- Copiando Ficheros...")
copy_tree(DEFAULT_PATH, PATH_71x90)
copy_tree(DEFAULT_PATH, PATH_35x45)
copy_tree(DEFAULT_PATH, PATH_17x22)
print("--- Ficheros Copiados")

#--- FIN

def resize_images(path, size):

    sub_paths = ['Train', 'Validation'] #-Airan: Sub_Path

    # Carpetas con muestras
    folders = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise' ]

    print("--- Cambiando tama√±o...")
    a=True
    for fex in folders:
        for sub_path in sub_paths: #-Airan: Obtener todo el dataset: Train, Validation

            folder = os.path.join((path + sub_path), fex)

            for file_name in os.listdir(folder):

                # Asume im√°genes en formato png
                if file_name.endswith('.png'):
                    
                    #-Airan: Ruta completa imagen
                    abs_path = os.path.join(folder, file_name)                  
                    
                    #-Airan: Lee la imagen
                    image = cv2.imread(abs_path)
                    
                    if a:
                        plt.imshow(image)
                        plt.show()
                    
                    #-Airan: Modificaci√≥n tama√±o imagen
                    image = cv2.resize(image, size) # Air√°n
                    
                    if a:
                        plt.imshow(image)
                        plt.show()
                        print(abs_path)
                        a=False
                    
                    cv2.imwrite(abs_path, image)
    print("--- Tama√±o cambiado.")

        
resize_images('./SFEW_71x90/', (71, 90))
resize_images('./SFEW_35x45/', (35, 45))
resize_images('./SFEW_17x22/', (17, 22))
```
